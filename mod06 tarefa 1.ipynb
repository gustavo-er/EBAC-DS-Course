```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('./SINASC_RO_2019.csv')
df.columns
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig1 = df[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
fig1['ESTCIVMAE'] = fig1['ESTCIVMAE'].replace([1, 2, 3, 4, 5, 9], ['Solteira', 'Casada', 'Viúva', 'Divorciado', 'Uniao Consensual', 'Ignorado'])
```

```python
sns.countplot(x='APGAR1', data=fig1)
```

```python
sns.countplot(x='APGAR5', data=fig1)
```

```python
sns.countplot(x='SEXO', data=fig1)
```

```python
sns.countplot(x='GESTACAO', data=fig1)
```

```python
sns.countplot(x='ESTCIVMAE', data=fig1).tick_params(axis='x', rotation = 90)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
df2 = df[['APGAR5', 'CONSULTAS']]
sns.barplot(x='APGAR5', y='CONSULTAS', estimator=sum, data=df2, ci = None)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

ax1 = sns.countplot(ax = axes[0, 0], x='SEXO', data=fig1)
ax2 = sns.countplot(ax = axes[0, 1], x='CONSULTAS', data=df)
ax3 = sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=fig1)
ax4 = sns.countplot(ax = axes[1, 1], x='GESTACAO', data=fig1)

ax3.tick_params(axis='x', rotation = 90)
ax4.tick_params(axis='x', rotation = 90)
```
